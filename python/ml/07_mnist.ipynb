```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
plt.matshow(digits.images[321])
```

```python
X, y = digits.data, digits.target
```

```python
X.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

X_test_predict = knn.predict(X_test)
print(X_test_predict.shape)
print(len(np.linspace(0,100, len(X_test_predict))))
```

```python
plt.plot(np.linspace(0, 100000000, len(X_test_predict)) , X_test_predict)
plt.plot(np.linspace(0, 100000000, len(X_test_predict)), y_test, color='r')
```
